```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from random import randrange
import random
from scipy.spatial.distance import cdist
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
# Create Data
data, labels = make_blobs(n_samples=200, n_features=2, 
                           centers=2, cluster_std=4.0, random_state=100)
x = data[:,0]
y = data[:,1]
```

```python
sns.set_style("whitegrid")
plt.scatter(x , y, c = labels, cmap='rainbow')
```

```python
import knn
```

```python
k_opt, accuracy = knn.cross_val(data, labels, method="class", n_bags=20, kmax=30, seed = 1)
```

```python
k_opt
```

```python
plt.plot(accuracy/100.0,'o-')
```

```python
pred_test=knn.testing(train = data, test=data, labels=labels, k_opt = 10)
```

```python
knn.getAccuracy(pred_test,labels)
```

## Regression

```python
n = 30
rand = np.random.rand(n)*2 - 1
x = np.array(list(range(n)))
y = 1.0 + 2.0*x + 4.5*rand
z = -2.0 + y + rand
```

```python
data = np.column_stack((x,y))
```

```python
plt.plot(x,z,'o-')
```

```python
k_opt, accuracy = knn.cross_val(data = data, labels=z, method="reg", n_bags=10, kmax=40, seed = 10)
accuracy
```

```python
plt.plot(1-accuracy)
```

```python
pred_test=knn.testing(train = data, test = data, labels=z, method="reg", k_opt = 25)
```
